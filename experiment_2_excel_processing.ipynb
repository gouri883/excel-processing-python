Reading XLS using xlrd


```python
import xlrd

workbook = xlrd.open_workbook(file_path1)
sheet = workbook.sheet_by_index(0)

value = sheet.cell_value(0, 0)
print(value, '\n')

for i in range(sheet.ncols):
    print(sheet.cell_value(3, i))
```

```python
row0 = sheet.row_values(0)
print("Row 0:", row0)

col1 = sheet.col_values(1)
print("Column 1:", col1)

cell_1_0 = sheet.cell_value(1, 0)
print(cell_1_0)
```

```python
num_rows = sheet.nrows
num_cols = sheet.ncols

print(f"Number of rows: {num_rows}")
print(f"Number of columns: {num_cols}")
```

```python
import xlwt

wb = xlwt.Workbook()
ws = wb.add_sheet('Sheet1')

row_data = ['First Name', 'Last Name', 'Age']
for c, val in enumerate(row_data):
    ws.write(0, c, val)

col_data = ['Alice', 'Bob', 'Charlie']
for r, val in enumerate(col_data, start=1):
    ws.write(r, 0, val)

col_data = ['Wonderland', 'Charlton', 'Chaplin']
for r, val in enumerate(col_data, start=1):
    ws.write(r, 1, val)

col_data = [15, 18, 23]
for r, val in enumerate(col_data, start=1):
    ws.write(r, 2, val)

wb.save('output.xls')
```

Read and Write using openpyxl (.XLSX)

```python
!pip install openpyxl
```

```python
from openpyxl import load_workbook


wb = load_workbook(file_path2)
sheet = wb.active


print("A1:", sheet["A1"].value)

row1 = [cell.value for cell in sheet[1]]
print("Row 1:", row1)

colA = [cell.value for cell in sheet['A']]
print("Column A:", colA[:10], "...")

value = sheet.cell(row=2, column=3).value
print("Cell (2,3):", value)
```

```python
for sheet_name in wb.sheetnames:
    sheet = wb[sheet_name]
    print("\n========================")
    print("Sheet:", sheet_name)
    print("========================")

    print("A1:", sheet["A1"].value)

    row1 = [cell.value for cell in sheet[1]]
    print("Row 1:", row1)

    colA = [cell.value for cell in sheet['A']]
    print("Column A (first 10):", colA[:10])

    value = sheet.cell(row=2, column=3).value
    print("Cell (2,3):", value)
```

```python
from openpyxl import load_workbook
sheet = wb.active

sheet["Z1"] = "New Column"

new_values = [10, 20, 30, 40, 50]

for i, val in enumerate(new_values, start=2):
    sheet.cell(row=i, column=26, value=val)

output_path = "/content/drive/MyDrive/SOWC_updated.xlsx"
wb.save(output_path)

print("Saved:", output_path)
```

Reading and Writing using Pandas(XLSX)

* Reading - Pandas
* Writing - Pandas + openpyxl

```python
import pandas as pd

file_path2 = '/content/drive/MyDrive/SOWC_2014_Stat_Tables_Table_9.xlsx'
dfs = pd.read_excel(file_path2, sheet_name=None)
for sheet_name in dfs:
    dfs[sheet_name]["sheets"] = sheet_name
combined = pd.concat(dfs.values())
combined[combined["sheets"] == "Table 9 "]

clean_dfs = {name: df for name, df in dfs.items() if df.dropna(how="all").shape[0] > 5}

for sheet_name, df in clean_dfs.items():
    print("\n=============================")
    print("SHEET NAME:", sheet_name)
    print("=============================")
    print(df.head(10))
    print("Shape:", df.shape)
```

```python
sheet_names = list(dfs.keys())
sheet1 = sheet_names[0]
sheet2 = sheet_names[1]

import pandas as pd

output_path = "/content/drive/MyDrive/output_two_sheets.xlsx"

with pd.ExcelWriter(output_path, engine="openpyxl") as writer:
    dfs[sheet1].to_excel(writer, sheet_name=sheet1, index=False)
    dfs[sheet2].to_excel(writer, sheet_name=sheet2, index=False)

print("Saved:", output_path)
```
